```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create dummy data
np.random.seed(42)
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': np.random.randint(1, 10, size=5)
})
```

```python
# Bar chart
plt.figure(figsize=(8, 4))
plt.bar(data['Category'], data['Value'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(data['Value'], labels=data['Category'], startangle=90, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Line chart
plt.figure(figsize=(8, 4))
plt.plot(data['Category'], data['Value'], marker='o')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Line Chart')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(data['Category'], data['Value'], marker='o', color='red')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Scatter Plot')
plt.show()
```
